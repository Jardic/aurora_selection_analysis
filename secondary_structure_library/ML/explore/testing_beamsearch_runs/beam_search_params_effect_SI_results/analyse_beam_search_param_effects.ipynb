```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'Arial'
```

```python
df = pd.read_csv('beam_search_params_effect_SI_results.csv', index_col=0)
del df['recall']
del df['top_explored']
```

```python
plt.figure(figsize=(3, 2.5))
sns.scatterplot(data=df, x='max_depth', y='precision', color='grey', edgecolor='grey', alpha=0.5)
sns.lineplot(data=df, x='max_depth', y='precision', color='grey')

plt.xlabel('Max depth')
plt.ylabel('Precision')

sns.despine()
plt.tight_layout()

plt.savefig('beam_search_params_max_depth_vs_precision.svg')
```

```python
plt.figure(figsize=(3, 2.5))
sns.scatterplot(data=df, x='max_depth', y='search_space', color='grey', edgecolor='grey', alpha=0.5)
sns.lineplot(data=df, x='max_depth', y='search_space', color='grey')

plt.xlabel('Max depth')
plt.ylabel('Search space')

sns.despine()
plt.tight_layout()

plt.savefig('beam_search_params_max_depth_vs_search_space.svg')
```
