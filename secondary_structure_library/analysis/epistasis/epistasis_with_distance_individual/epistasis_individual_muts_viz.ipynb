```python
for mut, df_m in results_final:

    mut = [tuple([dct_tr_pos[x] for x in mut[0]]), mut[1]]

    mutname_title = '-'.join([str(x) for x in mut[0]]) + '|' + 'â†’'.join([x for x in mut[1]])
    mutname_file = '-'.join([str(x) for x in mut[0]]) + '|' + '_'.join([x for x in mut[1]])
    df_m = df_m[df_m['d'].between(1, 6)]

    # Define absolute axis size in inches
    axis_width = 2  # Width in inches
    axis_height = 1.5  # Height in inches
    w = 2
    a = 0.3
    
    # Define margins
    left_margin = 0.5
    bottom_margin = 0.5
    
    # Compute figure size to ensure fixed axis size
    fig_width = left_margin + axis_width + 0.5
    fig_height = bottom_margin + axis_height + 0.5
    
    fig = plt.figure(figsize=(fig_width, fig_height))
    
    # Manually place the axes (left, bottom, width, height) in figure coordinates (0-1 range)
    ax = fig.add_axes([
        left_margin / fig_width,
        bottom_margin / fig_height,
        axis_width / fig_width,
        axis_height / fig_height
    ])

    sns.violinplot(data=df_m, x='d', y='log_diff_effect',
               bw_adjust=1,
               inner=None,
               linewidth=1,
              )
    
    #sns.violinplot(data=df_m, x='d', y='log_diff_effect', alpha=0.3, saturation=1, inner='point', inner_kws={'alpha':0.1, 'rasterized':True})
    #sns.stripplot(data=df_m, x='d', y='log_diff_effect', alpha=0.3, rasterized=True)
    #sns.boxplot(data=df_m, x='d', y='log_diff_effect', color='grey', fill=False)

    #sns.stripplot(data=df[df['background_distance'] <= 6], x='background_distance', y='avg_effect_ratio', alpha=0.01, jitter=0.1, color='tab:blue', rasterized=True)
    #sns.boxplot(data=df[df['background_distance'] <= 6], x='background_distance', y='avg_effect_ratio', color='grey', fill=False)
    
    #for i in range(-4, 5):
    #    plt.axhline(i, alpha=0.4, color='grey', linewidth=0.3)
    
    plt.title(mutname_title)
    
    plt.ylim(-4.5, 4.5)
    
    #plt.xticks([])
    #plt.yticks([0, 1, 2, 3])
    
    plt.xlabel('Background distance')
    plt.ylabel('Log fold diff.')

    plt.grid(alpha=0.1)
    
    sns.despine()
    plt.savefig('mutation_' + mutname_file + '.svg', dpi=300)
    plt.close()
```
