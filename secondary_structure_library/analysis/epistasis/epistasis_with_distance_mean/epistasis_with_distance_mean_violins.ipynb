```python
df = pd.read_csv('epistasis_with_distance_ref50.csv', index_col=0)
df = df[df['background_distance'].between(1, 6)]

# Define absolute axis size in inches
axis_width = 2.2  # Width in inches
axis_height = 1.5  # Height in inches
w = 2
a = 0.3

# Define margins
left_margin = 0.5
bottom_margin = 0.5

# Compute figure size to ensure fixed axis size
fig_width = left_margin + axis_width + 0.5
fig_height = bottom_margin + axis_height + 0.5

fig = plt.figure(figsize=(fig_width, fig_height))

# Manually place the axes (left, bottom, width, height) in figure coordinates (0-1 range)
ax = fig.add_axes([
    left_margin / fig_width,
    bottom_margin / fig_height,
    axis_width / fig_width,
    axis_height / fig_height
])

sns.violinplot(data=df, x='background_distance', y='avg_effect_ratio',
               bw_adjust=1,
               inner='quart',
               linewidth=1
              )

plt.yticks([-2, 0, 2, 4])
plt.ylim(-2, 5)

plt.xlabel('Background distance')
plt.ylabel('Avg. log fold diff.')

sns.despine()
plt.savefig('avg_mut_effect_ratio_vs_background_dist_singles_ref_50.svg')
```
