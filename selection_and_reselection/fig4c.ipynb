# Number of sequences vs. numer of reads plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Matplotlib settings
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['font.size'] = 13
plt.rcParams['font.family'] = 'Arial'

yellow = '#ffc107'
purple = '#303f9f'
```

```python
# Reading datasets
df_init = pd.read_csv('/home/kurfurst/Documents/datasets/MV_aurora_singlestep_initial.tsv', sep='\t', names=['count', 'seq'])
df_select = pd.read_csv('/home/kurfurst/Documents/datasets/MV_aurora_singlestep_10min_20uM.tsv', sep='\t', names=['count', 'seq'])
```

```python
# Value counts of the readnumbers
vc_select = df_select['count'].value_counts().sort_index()
vc_init = df_init['count'].value_counts().sort_index()
```

```python
w = 1

plt.bar(vc_select.index, vc_select.values, width=w, color=yellow, label='Selected pool')
plt.bar(vc_init.index, vc_init.values, width=w, color=purple, label='Initial pool')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Number of reads')
plt.ylabel('Number of sequences')
plt.legend()
plt.tight_layout()
#plt.show()
plt.savefig('fig4b_10min_20uM.png', dpi=180)
```
